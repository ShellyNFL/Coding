```python
import matplotlib.pyplot as plt
import numpy as np #hace cálculos numéricos
x=np.arange(-10,10,0.1)
c1=1 # solo piden que sea valor positivo
y1=c1*x-x*np.cos(x)

#-----------------
c2=0
y2=c2*x-x*np.cos(x) #para llamar la función
#-----------------
c3=-1
y3=c3*x-x*np.cos(x)

plt.plot(x,y1, '-r', label= 'y1, c>0')

plt.plot(x,y2, ':b', label= 'y2, c=0')

plt.plot(x,y3, '--k', label= 'y3, c<0')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Lab DLS')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
